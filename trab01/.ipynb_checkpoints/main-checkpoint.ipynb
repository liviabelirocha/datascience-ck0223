## Passo 1
Importar libs

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Passo 2
Carregar csv

```python
data = pd.read_csv('csv/marvel_demo_stats_powers.csv')

data
```

Remoção de colunas inúteis

```python
data = data.drop(columns=["Unnamed: 0"])
```

# Passo 3
Plotar algumas informações sobre o dataset.

```python
data.columns.values.tolist()
```

```python
data.info(verbose=True, show_counts=True)
```

```python
plt.figure(figsize=(12,12))
plt.title("Dados faltantes", size=30)
sns.heatmap(data.isna());
```

```python
percentage = data['Gender'].value_counts()/data['Gender'].count() * 100;
print(percentage)

data['Gender'].value_counts().plot(kind='bar');
```

Pode-se ver que aproximadamente 64% dos personagens são homens, 29% são mulheres e 6% não possuem gênero.
